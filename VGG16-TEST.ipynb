```
def predict_picture(img_path):
  # 画像を指定パスからロードする。
  # 同時にネットワークが受け付けることのできる形にリサイズする。
  img = keras.preprocessing.image.load_img(img_path, target_size=(224, 224))

  # 対象画像を表示
  plt.figure()
  plt.imshow(img)
  # ax.grid(color='w', linestyle='none')


  # 画像をnumpy配列に変換し、さらに4階テンソルへ変換
  x = keras.preprocessing.image.img_to_array(img)
  x = np.expand_dims(x, axis=0)

  # VGG16用の前処理。
  x = keras.applications.vgg16.preprocess_input(x)

  # クラスを予測
  # pred()は実は複数枚うけとって、複数の結果を返す。
  preds = model.predict(x)
      
  # 入力は1枚の画像を含む4階テンソルなので、出力から[0]のみとりだす。
  results = keras.applications.vgg16.decode_predictions(preds, top=5)[0]
  
  # VGG16の1000クラスはdecode_predictions()で文字列に変換される
  for result in results:
    print(result)
  print()
```

```
#ディレクトリにあるすべての画像を推定してみます。
```

```
import glob
files = glob.glob("*.png")
```

```
for file in files:
  predict_picture(file)
```
